```python
model_list = ["Site_independent", "EVmutation",
              "EVE", "ESM1v", 'ESM2', 'Progen2',
             'Tranception_no_retrieval', 
              'MSA_Transformer', 'PoET', 'ProteinMPNN', 'ESM-IF1', 
             'SaProt_650M_AF2', 'METL']

base_dir = "/n/groups/marks/users/sam/SpecificityStudio/model_scores"

spec_studio_reference_file = pd.read_csv("/n/groups/marks/users/sam/SpecificityStudio/SpecificityStudio_reference_sheet.csv", index_col=0)

all_data = {}
for DMS_id in spec_studio_reference_file.index:

    if DMS_id!='POLG_TEV_Huber_2025' and DMS_id!='TRPB1_PYRFU_Zhoufan_Li_2025':
        print(f'Assembling scores for dataset {DMS_id}')
        dataset = pd.read_csv(f'../processed_data/{DMS_id}.csv')
        dataset['mutant'] = dataset['mutant'].replace('M1M', 'WT').replace('G1G', 'WT').replace('wt', 'WT')

        #print(dataset.loc[dataset.mutant=='M1M'])
        
        reference_seq = spec_studio_reference_file.loc[DMS_id].target_seq
        model_predictions = {}
        for model in model_list:
        
            try:
                model_predictions[model] = pd.read_csv(f"../model_scores/{model}/{DMS_id}.csv")
            except:
                print(f"Mising model scores for {DMS_id} with {model}")

        merged_df = merge_model_scores(
            data_root=base_dir,
            model_score_csv="ss_model_score_names.csv",
            reference_seq=reference_seq,
            assay_filename=f"{DMS_id}.csv",
            output_csv=f"{DMS_id}_merged_model_scores.csv"
        )

        merged_df = merged_df.loc[[len(i)>0 for i in merged_df['mutant']]]
        merged_df['mutant'] = merged_df['mutant'].replace('M1M', 'WT').replace('G1G', 'WT').replace('wt', 'WT')
        merged_df = consolidate_duplicate_indices(merged_df.set_index('mutant'))

        dataset = consolidate_duplicate_indices(dataset.set_index("mutant"))
        
        full_df_with_model_scores = pd.concat([dataset, merged_df], axis=1).dropna(thresh=1)

        full_df_with_model_scores.to_csv(f"../model_scores/Aggregated/{DMS_id}_scores_v1.csv")
        
        all_data[DMS_id] = full_df_with_model_scores

        # except:
        #     vals, counts = np.unique(dataset.mutant, return_counts=True)
        #     print(pd.Series(counts, index=vals).sort_values(ascending=False))
            
        print('')
```

```python
import pickle
with open("all_scores_Jun2026_v1.pkl", "wb") as outfile:
    pickle.dump(all_data, outfile)
```
